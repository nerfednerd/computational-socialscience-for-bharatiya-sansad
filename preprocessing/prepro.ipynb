```python
import os

df=pd.read_csv(input_file)
print(list(df["Speaker"]))
```

```python
# To tackle the constituency based inconsistencies -> remove parantheses and everything inside them
import re
def remove_para_content(name:str)-> str:
    name = re.sub(r'\(.*?\)', '', name)
    return name
df["Speaker"] = df["Speaker"].apply(remove_para_content)
print(list(df["Speaker"]))
```

```python
import re
def remove_long_pseudo_names(name:str)-> str:
    if not isinstance(name, str):
        return name
    if re.search(r"[A-Za-z]", name):
        name=name.upper()
    if len(name)>30:
        return None
    return name

df["Speaker"] = df["Speaker"].apply(remove_long_pseudo_names)
print(list(df["Speaker"]))

```

```python
# Remove the first word, the title/honorofic
# Remove any puntunation in between the name
# Trim Extra Spaces
# Remove and zero width characters like \u200b etc.

def name_normalization_one(name: str)->str:
    if pd.isna(name):
        return name
    name = re.sub(r'[\u200b\u200c\u200d]', '', name)
    name = re.sub(r'\(.*?\)', '', name)
    name = re.sub(r'^\S+\s*', '', name)
    name = re.sub(r'[.\-]', '', name)
    name = name.strip()
    return name

df["Speaker"] = df["Speaker"].astype(str).apply(name_normalization_one)
display(df)
```

```python
df = df[~((df['Words_in_Speech'] < 50) | (df['Speaker'].isna()) | (df['Speaker'].str.strip() == ''))]
df = df.reset_index(drop=True)

pd.set_option('display.max_rows', None)
display(df)

```

```python
import pandas as pd
import re
from aksharamukha import transliterate

# detect if text contains Devanagari characters
DEVANAGARI_RE = re.compile(r'[\u0900-\u097F]')

def transliterate_if_hindi(name, lang):
    if pd.isna(name) or not isinstance(name, str) or name.strip() == '':
        return name

    lang_str = '' if pd.isna(lang) else str(lang).strip().lower()
    is_hindi = (lang_str in ('hi', 'hindi')) or bool(DEVANAGARI_RE.search(name))

    if not is_hindi:
        return name

    try:
        # Convert from Devanagari → Latin (ISO) and make uppercase
        transliterated = transliterate.process('Devanagari', 'ISO', name)
        return transliterated.upper()
    except Exception:
        return name  # fallback if transliteration fails

# Apply to your dataframe
df['Speaker'] = df.apply(lambda row: transliterate_if_hindi(row['Speaker'], row.get('Language')), axis=1)

# See result
display(df)
```

```python
from rapidfuzz import fuzz
from collections import defaultdict

def normalize_name(name):
    return ' '.join(name.upper().split())

def group_names(names, threshold_main=80, threshold_secondary=70):
    names = [normalize_name(n) for n in names]

    # Step 1: Build initial graph based on main threshold
    graph = defaultdict(list)
    for i in range(len(names)):
        for j in range(i + 1, len(names)):
            sim = fuzz.token_set_ratio(names[i], names[j])
            if sim >= threshold_main:
                graph[names[i]].append(names[j])
                graph[names[j]].append(names[i])

    # Step 2: Assign initial groups using connected components
    visited = set()
    id_counter = 1
    groups = {}
    for name in names:
        if name not in visited:
            stack = [name]
            component = []
            while stack:
                current = stack.pop()
                if current not in visited:
                    visited.add(current)
                    component.append(current)
                    for neighbor in graph[current]:
                        if neighbor not in visited:
                            stack.append(neighbor)
            for c in sorted(component):
                groups[c] = id_counter
            id_counter += 1

    # Step 3: Secondary pass for “fuzzy first/last name matches”
    group_items = list(groups.items())
    for i in range(len(group_items)):
        name1, id1 = group_items[i]
        words1 = set(name1.split())
        for j in range(i + 1, len(group_items)):
            name2, id2 = group_items[j]
            if id1 != id2:
                words2 = set(name2.split())
                overlap = len(words1 & words2)
                total = max(len(words1), len(words2))
                partial_ratio = (overlap / total) * 100
                if partial_ratio >= threshold_secondary:
                    # Merge groups
                    old_id, new_id = id2, id1
                    for k in groups:
                        if groups[k] == old_id:
                            groups[k] = new_id

    return groups

final_names=df["Speaker"].unique()
groups = group_names(final_names, threshold_main=70, threshold_secondary=80)
print(groups)
```

```python
import pandas as pd

# Assuming your DataFrame is df

# Aggregate
agg_df = df.groupby('Speaker').agg({
    'Speech': lambda x: ' <u> '.join(x),      # concatenate speeches
    'Words_in_Speech': 'sum',                # sum words
    'Language': 'first'                      # keep language as it is
}).reset_index()

display(agg_df)
```

```python
import pandas as pd

# Assuming df is your DataFrame and groups is your dictionary
# Step 1: Map each speaker to their group
df['group_id'] = df['Speaker'].map(groups)

# Step 2: Aggregate by group_id
agg_df = df.groupby('group_id').agg({
    'Speaker': lambda x: ' | '.join(sorted(set(x))),   # all speaker variants
    'Speech': lambda x: ' <u> '.join(x),              # all speeches
    'Words_in_Speech': 'sum',                         # total words
    'Language': 'first'                               # assuming same language per group
}).reset_index(drop=True)

# Display the aggregated result
display(agg_df)
```
