```python
import re
def clean_name_16(name):
    if pd.isna(name):
        return name
    
    #remove anything enclosed within parantheses
    name = re.sub(r'\(.*?\)', '', name)
    honorifics = ['Shri', 'Prof', 'Dr', 'Shrimate', 'Prof.', 'Dr.', 'Smt.']
    parts=name.split()
    cleaned_parts=[part for part in parts if part not in honorifics]
    return " ".join(cleaned_parts)
df_16["Name"]=df_16["Name"].apply(clean_name_16)
df_16
```

```python
df_18=pd.read_excel("SittingMembers_18LokSabha.xlsx")
df_18
```

```python
df_18=df_18[["Name of Member", "Party Name"]]
df_18
```

```python
df_18=df_18.rename(columns={
    "Name of Member":"Name",
})
df_18
```

```python
def clean_name_18(name):
    if pd.isna(name):
        return name
    name = re.sub(r'\(.*?\)', '', name)
    honorifics = [
        'shri', 'shrimate', 'smt.', 'mr', 'mrs', 'ms', 'prof', 'dr', 
        'prof.', 'dr.'
    ]
    if ',' in name:
        last, first = name.split(',', 1)
        first_parts = [part for part in first.split() if part.lower() not in honorifics]
        first_cleaned = " ".join(first_parts)
        # Combine first and last
        return f"{first_cleaned.strip()} {last.strip()}"
    else:
        # If no comma, just clean honorifics
        parts = [part for part in name.split() if part.lower() not in honorifics]
        return " ".join(parts).strip()
df_18['Name'] = df_18['Name'].apply(clean_name_18)
df_18
```

```python
import pandas as pd

df_17 = pd.read_csv("lok_sabha_all_members_17.csv")

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)

mask = df_17["Name"].fillna("").str.strip().str.lower() != "vacant"
df_17 = df_17[mask]
df_17 = df_17.dropna(subset=["Name"])

cols_to_remove_17=["#", "Constituency", "Party"]
df_17=df_17.drop(columns=cols_to_remove_17)

df_17=df_17.rename(columns={
    "Party.1":"Party Name"
})


df_17["Party Name"] = df_17["Party Name"].ffill()


#to remove content inside the paratheses, which primarily contains information about people who switched parties, resigned or passed away. 
#for now we are assuming that people did not switch parties in between the session, as the numbers of people switching parties in the lok sabha are comparatively lower.
def remove_paranthesis_17(name):
    if pd.isna(name):
        return name
    name = re.sub(r'\(.*?\)', '', name)
    return name

df_17["Name"]=df_17["Name"].apply(remove_paranthesis_17)
df_17 = df_17[~df_17["Name"].str.contains("party" or "congress", case=False, na=False)]


display(df_17)



```
